```python
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import brown

dataset=[]
```

```python
for category in brown.categories():
    for fileid in brown.fileids(category):
        dataset.append((brown.words(fileids = fileid),category))
        

dataset = [[w.lower() for w in text] for text,category in dataset]
```

```python
dataset[1]
```

```python
joined_text = []
for para in dataset:
    joined_text.append(' '.join([word for word in para]))
    
joined_text[0]
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
import re

tokenizer = RegexpTokenizer(r'\w+')

stop_words = set(stopwords.words('english'))


def textclean(text):
    tokens = tokenizer.tokenize(text)
    tokens = [word for word in tokens if word.isalpha()]
    tokens = [word for word in tokens if not word in stop_words]
    tokens = [word for word in tokens if len(word) > 1]
    return ' '.join(tokens)
```

```python
from nltk.tokenize import sent_tokenize

doc_df = pd.DataFrame()

sentences = []
doc_id = []
count = 0

for para in joined_text:
    para_sentences = sent_tokenize(para)
    cleaned_sentences = [textclean(sent) for sent in para_sentences]
    sentences.extend(cleaned_sentences)
    doc_id.extend([count for i in range(len(cleaned_sentences))])
    count += 1
len(sentences[0])
```

```python
def generate_ctf_table(l):
    concepts = dict()
    for vas_dicts in l:
        for v in vas_dicts.values():
            if v not in concepts:
                concepts[v] = 1
            else:
                concepts[v] +=1
    return concepts
```

```python
def get_tf(concept,sent):
    concept_cnt = 0
    if concept in sent:
        concept_cnt += 1
    return concept_cnt
```

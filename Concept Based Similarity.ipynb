```python
def extractconcepts(l):
    concepts = []
    for vas_dicts in l:
        for v in vas_dicts.values():
            if v not in concepts:
                concepts.append(v)
    return concepts
```

```python
def get_ctf(concept,l):
    concept_cnt = 0
    for vas_dicts in l:
        for v in vas_dicts.values():
            if v == concept:
                concept_cnt += 1
    return concept_cnt
```

```python
def get_tf(concept,l):
    concept_cnt = 0
    for sentence in l:
        if concept in sentence:
            concept_cnt += 1
    return concept_cnt
```
